```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the paths to the image and mask directories
image_dir = r"D:\Holland_Year_2\Block_B\Task5_Patches\dataset_\train_images\train"
mask_dir = r"D:\Holland_Year_2\Block_B\Task5_Patches\dataset_\train_masks\train"

# Get the list of all image files in the image directory
image_files = [f for f in os.listdir(image_dir) if f.endswith('.png')]

# Select a random image file
random_image_file = random.choice(image_files)

# Construct the full path to the selected image and its corresponding mask
image_path = os.path.join(image_dir, random_image_file)
root_mask_path = os.path.join(mask_dir, random_image_file.replace('.png', '.tif'))

# Read the image and mask
im = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
root_mask = cv2.imread(root_mask_path, cv2.IMREAD_GRAYSCALE)

# Check if the image and mask were read correctly
if im is None:
    raise FileNotFoundError(f"Image file not found: {image_path}")
if root_mask is None:
    raise FileNotFoundError(f"Mask file not found: {root_mask_path}")

# Ensure the image and mask have the same shape
assert(im.shape == root_mask.shape)

# Ensure the mask contains only binary values (0 or 1)
assert(np.all(np.logical_or(root_mask == 0, root_mask == 1)))

# Ensure the mask has more background (0) than foreground (1)
assert(np.sum(root_mask == 0) > np.sum(root_mask == 1))

# Create an overlay of the mask on the image
root_color = [1, 0, 0]  # Red for root mask
im_rgb = np.dstack([im, im, im])
overlay_root = np.zeros_like(im_rgb, dtype=float)
alpha = 0.4  # Transparency level
overlay_root[root_mask == 1] = root_color

im_overlay = im_rgb / 255.0  # Normalize image
im_overlay += alpha * overlay_root
im_overlay = np.clip(im_overlay, 0, 1)  # Clip values to keep them within [0, 1]

# Plot the original image and overlay in the same figure
plt.figure(dpi=300)
plt.subplot(1, 2, 1)
plt.imshow(im, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(im_overlay)
plt.title('Root Mask Overlay')
plt.axis('off')

plt.suptitle(f'Image and Overlay: {random_image_file}')
plt.show()
```
